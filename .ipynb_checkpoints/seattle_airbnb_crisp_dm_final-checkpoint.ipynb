# Three Questions:
Part I: Which neighborhood is most budget-friendly for the stay in Seattle while maintaining a good experience for your time?\
Part II: What are features that could have a direct impact on the price and cleaning cost of Airbnb?\
Part III: What features would be the top indicators and drivers on the price of an Airbnb?

# Part I

```python
q1 = listings[['neighbourhood_group_cleansed',
               'review_scores_rating',
               'accommodates',
               'price',
               'cleaning_fee']]
q1.dtypes
```

```python
q1.isnull().sum()
```

```python
np.sum(q1['cleaning_fee']=='$0.00')
```

It is reasonable to fill cleaning_fee with 0, since all the cleaning fee doesn't have 0.\
Meanwhile, for review_scores_rating, there is a high likelihood that the rating is missing\
because they are new houses that haven't been rated yet. So they should not be filled with\
imputed values

Fill null cleaning fee as 0 and transform all prices and cleaning fees to float

```python
q1['cleaning_fee'].fillna('$0',inplace=True)
```

```python
q1[q1.columns[3:]] = q1[q1.columns[3:]].apply(lambda x: x.str.replace('$','')\
                                             ).apply(lambda x: x.str.replace(',','')\
                                                    ).astype(float)
```

Doing aggregation and metrics calculation

```python
q1.dropna(axis=0,how='any')
q1['price_per_person']=(q1['price']+q1['cleaning_fee'])/q1['accommodates']
q1_agg=q1.groupby('neighbourhood_group_cleansed').mean()
q1_agg['rating/price']=q1_agg['review_scores_rating']/q1_agg['price_per_person']
q1_agg=q1_agg.sort_values('rating/price',ascending=False)
q1_agg
```

Plot bar charts

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
neighbourhoods=q1_agg.index
rating_over_price=q1_agg['rating/price']
ax.bar(neighbourhoods,rating_over_price)
plt.xticks(rotation=75)
plt.title('Ratings over Price for each neighbourhood')
plt.show()
```

# Part II

Cleaning prices and cleaning fees following part I method,\
then selecting the subset for analysis,\
and adding new features indicating having square feet or not due to sparse of data.

```python
q2=listings.copy()
q2[q2.columns[q2.columns=='price']] = q2[q2.columns[q2.columns=='price']\
                                        ].apply(lambda x: x.str.replace('$','')\
                                               ).apply(lambda x: x.str.replace(',','')\
                                                      ).astype(float)
q2['cleaning_fee'].fillna('$0',inplace=True)
q2[q2.columns[q2.columns=='cleaning_fee']] = q2[q2.columns[q2.columns=='cleaning_fee']\
                                               ].apply(lambda x: x.str.replace('$','')\
                                                      ).apply(lambda x: x.str.replace(',','')\
                                                             ).astype(float)
q2.select_dtypes(exclude=['object']).columns
```

```python
q2_subset=q2[['host_total_listings_count','accommodates','bathrooms','bedrooms','beds',\
                  'square_feet','price','cleaning_fee','guests_included','number_of_reviews','review_scores_rating']]
q2_subset.describe()
```

```python
q2_subset['has_square_feet']=np.where(q2_subset['square_feet'].isnull(),0,1)
```

Plot heatmap for correlation

```python
sns.set(rc = {'figure.figsize':(15,8)})

ax=sns.heatmap(q2_subset.corr(),annot=True,fmt='.1f')
ax.set_ylim(len(q2_subset.columns)+0.5, -0.5)
ax
```

# Part III

Clean part II results so that the data could be fed to model.

```python
q2_model=q2_subset.drop(columns=['square_feet'],axis=1)
q2_cleaned=q2_model.dropna(axis=0)
q2_cleaned.describe()
```

Remove high correlation columns and split data to train and test

```python
X=q2_cleaned.drop(columns=['price','bedrooms','beds'], axis=1)
y=q2_cleaned['price']
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=42, test_size=0.3)
```

Fit model

```python
lm=LinearRegression(normalize=True)
lm.fit(train_X,train_y)
test_y_preds=lm.predict(test_X)
train_y_preds=lm.predict(train_X)

"The r-squared score for train model is {} and {} for test_model.".\
format(r2_score(train_y, train_y_preds), r2_score(test_y, test_y_preds))
```

Get coefs

```python
coefs_df=pd.DataFrame()
coefs_df['est_int']=train_X.columns
coefs_df['coefs']=lm.coef_
coefs_df['abs_coefs']=np.abs(lm.coef_)
coefs_df=coefs_df.sort_values('abs_coefs',ascending=False)
coefs_df
```

Additional analysis on p-value significance of each coefs

```python
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

# Part IV Unused Analysis

I did some additional analysis on availability of listings for each months\
and then learned from data that the information is not telling anything.

```python
calendar['year']=pd.DatetimeIndex(calendar['date']).year
calendar['month']=pd.DatetimeIndex(calendar['date']).month
calendar_group1=calendar.groupby(['year','month']).size().\
                        to_frame('total_count').reset_index()
calendar_group2=calendar[calendar['available']=='t'].groupby(['year','month']).\
                        size().to_frame('count').reset_index()
calendar_total=pd.merge(calendar_group1, calendar_group2,how='inner', \
                       left_on=['year','month'],\
                        right_on=['year','month']\
                       )[['year','month','count','total_count']]
calendar_total['ratio']=calendar_total['count']/calendar_total['total_count']
calendar_total.sort_values('ratio',ascending=False)
```
